```python
# 1. Write a program to add multiple elements into a set.
my_set = {3,4,5,6}
elements_to_add = {7,8,9}
my_set.update(elements_to_add)
print("The new set is =",my_set)
```

```python
# 2. Write a program to remove an element from a set if it is present.

my_set = {3,4,5,6,"naila"}
element_str = input("Enter an element want to remove")
try :
    element = int(element_str)
except ValueError:
    element = element_str
if element in my_set:
        my_set.remove(element)
        print("After removed set is:",my_set)
else: 
    print("Element not found in the set")
    
```

```python
# 3. Write a program to find the union, intersection, and difference of two sets.

set1 = {1,2,3,4,5}
set2 = {6,7,8,9,3,4,5}

union = set1 | set2
print("The union of this two set is:",union)
intersection = set1 & set2
print("The intersection of this two set is:",intersection) 
difference = set1 - set2
print("Difference of this two set is:",difference)
```

```python
# 4. Write a program to check if one set is a subset of another. 

set1 = {1,2,3,4}
set2 = {1,2,3,4,5,6}
print(set1.issubset(set2))
print(set2.issubset(set1))
```

```python
# 5. Write a program to find all elements that are present in exactly one of the two sets.

a = {1,2,3,4,5}
b = {3,4,5,6,7}

print(a.symmetric_difference(b))
```
