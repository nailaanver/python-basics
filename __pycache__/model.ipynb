```python
# Create a Text Analyzer that counts words, characters, and sentences in a paragraph.

import re
def analyze_paragraph(paragraph):
    char_count = len(paragraph)
    word = paragraph.split()
    word_count = len(word)
    sentence_count = len(re.findall(r'[.!?]+[\"\')\]]*\s*',paragraph))
    if sentence_count == 0 and char_count > 0:
        sentence_count = 1
    return {
    "characters": char_count,
    "words": word_count,
    "sentences": sentence_count
}
paragraph_text = "This is an example paragraph. It contains multiple sentences! How many characters and words are there?"
analyze_result = analyze_paragraph(paragraph_text)
print("Character count: ",analyze_result['characters'])
print("Word count :",analyze_result['words'])
print("Sentence count:",analyze_result['sentences'])


    
```
