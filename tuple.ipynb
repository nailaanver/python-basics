```python
# 1. Write a program to find the maximum and minimum values in a tuple.

tuple = (5,3,7,9,2)
largest = max(tuple)
print("Maximum Value is ",largest)
smallest = min(tuple)
print("Maximum Value is ",smallest)
```

```python
# 2. Write a program to check if a given element exists in a tuple.

tuple = (5,3,7,9,2,"naila","i")
element_str = input("Enter an element:").lower()
try:
    element = int(element_str)
except ValueError:
    element = element_str

if element in tuple:
    print( element,"is exist in the tuple") 

else:
    print("not presenting in the tuple")
```
