```python
# Write a function to find the grade of a student from marks.

def find_grade(marks):
    if marks >= 90:
        return 'A'
    elif marks >= 80:
        return 'B'
    elif marks >= 70:
        return 'C'
    elif marks >= 60:
        return 'D'
    else:
        return 'F'
find_grade(198)

```

```python
# Write a function to check if a given year is a leap year.

def check_year(year):
    if (year % 4 ==0 and year % 100 !=0 ) or (year % 400 ==0):
        print(year, "is leap year")
    else:
        print(year, "is not a leap year")
check_year(2024)
```

```python
# Write a function that accepts a list of numbers and returns only the even numbers.

def check_even(list):
    even_numbers = []
    for i in list:
        if i%2 == 0:
           even_numbers.append(i)
    return even_numbers
list = [4,3,2,5,6]
check_even(list)
```

```python
# Write a function to calculate simple interest.

def calculate_simple_interest(principle,rate,time):
    simple_interest = (principle*rate*time)/100
    return simple_interest
calculate_simple_interest(1000,5,2)
```

```python
# Write a function that simulates a calculator (add, subtract, multiply, divide).

def calculator(a,b,operation):
    if operation == "add":
        return a+b
    elif operation =="subtract":
        return a-b
    elif operation == "multiply":
        return a*b
    elif operation == "divide":
        if b !=0:
            return a/b
        else:
            return "Error: Division by zero"
    else:
        "Error:Invalid operation"
        
calculator(2,3,"multiply")
```
