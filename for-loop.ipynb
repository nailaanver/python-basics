```python
#2. Write a program to print the multiplication table of a given number. 

num = int (input("Enter a number:"))
print("Multiplication table of ",num,"is:")
for i in range(1,11):
    result = i*num
    print(i,"*",num ,"=",result)
    i +=1
```

```python
#3. Write a program to print all even numbers from 1 to 50. 

for i in range(1,51):
    if i % 2 == 0:
        print(i)
```

```python
#4. Write a program to calculate the sum of numbers from 1 to 100 using a for loop. 

sum = 0
for i in range(1,101):
    sum +=i
print(sum)
```

```python
#5. Write a program to print the square of numbers from 1 to 10. 

square = 0
for i in range(1,11):
    square = i*i
    print("Square of",i,"is",square)
```
